# Part 1 Creating and Manipulating Arrays

## 1

```python
import numpy as np
from matplotlib import pyplot as plt
```

## 1.1

```python
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
x,y = np.meshgrid(x,y)
x,y
```

## 1.2

```python
plt.pcolormesh(x)
```

```python
plt.pcolormesh(y)
```

## 1.3

```python
radial = np.sqrt(x**2 + y**2)
angle = np.arctan2(y,x)
```

## 1.4

```python
plt.pcolormesh(x,y,radial)
```

```python
plt.pcolormesh(x,y,angle)
```

## 1.5

```python
f = (np.cos(4 * radial))**2 + (np.sin(4 * angle))**2
plt.pcolormesh(x,y,f)
```

## 1.6

```python
plt.plot(f.mean(axis = 1),y);
```

## 1.7

```python
plt.plot(x, f.mean(axis = 0));
```

The data covers 78 levels of depth recorded over 75 days. The ARGO Floats measure Temperature(T), Salinity(S), and Pressure(P), for each day and each level, which leads to a data frame with 78 rows and 75 columns where each day is a different column and each level is a different row. The profile has one latitude and one longitude for each day, which is why there are 75 measurements. This is also why there are 75 dates (one for each day). There are 78 levels, which is why levels have 78 entries.

## 2.3

```python
plt.plot(S,levels);
plt.ylim(80, 0)
```

```python
plt.plot(P, levels);
plt.ylim(80, 0)
```

## 2.4

```python
T_mean = T.mean(1)
S_mean = S.mean(1)
P_mean = P.mean(1)
T_std = T.std(1)
S_std = S.std(1)
P_std = P.std(1)
```

## 2.5

```python
plt.plot(T_mean, levels)
plt.ylim(80, 0)
plt.errorbar(T_mean, levels, xerr = T_std)
```

```python
plt.plot(S_mean, levels)
plt.ylim(80, 0)
plt.errorbar(S_mean, levels, xerr = S_std)
```

```python
plt.plot(P_mean, levels)
plt.ylim(80, 0)
plt.errorbar(P_mean, levels, xerr = P_std)
```

## 2.6

```python
T_nanmean = np.nanmean(T, axis = 1)
S_nanmean = np.nanmean(S, axis = 1)
P_nanmean = np.nanmean(P, axis = 1)
T_nanstd = np.nanstd(T, axis = 1)
S_nanstd = np.nanstd(S, axis = 1)
P_nanstd = np.nanstd(P, axis = 1)
plt.plot(T_nanmean, levels)
plt.ylim(80, 0)
plt.errorbar(T_nanmean, levels, xerr = T_nanstd)
```

```python
plt.plot(S_nanmean, levels)
plt.ylim(80, 0)
plt.errorbar(S_nanmean, levels, xerr = S_nanstd)
```

```python
plt.plot(P_nanmean, levels)
plt.ylim(80, 0)
plt.errorbar(P_nanmean, levels, xerr = P_nanstd)
```
